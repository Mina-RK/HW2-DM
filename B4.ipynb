```python
from sklearn.datasets import load_iris
import pandas as pd
data = load_iris()
from scipy.cluster.hierarchy import linkage, dendrogram
iris_data = pd.DataFrame(data.data, columns=data.feature_names)
new_iris = linkage(iris_data, method='complete')
new_iris
```

```python
dendrogram(new_iris,labels=None,leaf_rotation=90)
```

```python
from scipy.cluster.hierarchy import fcluster
num=fcluster(new_iris, criterion='maxclust', t=6)
num
```

```python
import numpy as np
import matplotlib.pyplot as plt
num=fcluster(new_iris, criterion='maxclust', t=6)
values, counts = np.unique(num, return_counts=True)
plt.scatter(values, counts)
```

```python
num=fcluster(new_iris, criterion='maxclust', t=5)
values, counts = np.unique(num, return_counts=True)
plt.scatter(values, counts)
```
