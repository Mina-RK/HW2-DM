```python
from sklearn import datasets
import pandas
brst_cncr = datasets.load_breast_cancer()
brst_cncr_dataframe = pandas.DataFrame(brst_cncr.data,columns=brst_cncr.feature_names)
brst_cncr_dataframe['target'] = pandas.Series(brst_cncr.target)
brst_cncr_dataframe.head()
```

```python
counts = brst_cncr_dataframe.target.value_counts(ascending=True)
counts
```

```python
X = brst_cncr_dataframe[brst_cncr_dataframe.columns[:-1]]
y = brst_cncr_dataframe.target
```

```python
from sklearn.model_selection import train_test_split
train_test_split(X, y, test_size=0.2)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
train_accuracy=[]
test_accuracy=[] 
for i in range (0, 10):
        model = KNeighborsClassifier(n_neighbors=i+1).fit(X_train, y_train)
        train_accuracy.append(model.score(X_train, y_train))
        test_accuracy.append(model.score(X_test, y_test))
```

```python
train_accuracy
```

```python
test_accuracy
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.scatter(train_accuracy, test_accuracy)
plt.show()
```

```python
plt.title('knn method')
plt.plot(np.arange(0,10),train_accuracy, label='train accuracy')
plt.plot(np.arange(0,10),test_accuracy, label='test accuracy')
plt.legend()
plt.xlabel('num of K')
plt.ylabel('accuracy')
plt.show()
```
