```python
from sklearn.datasets import load_iris
import pandas as pd
data = load_iris()
```

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
X=data.data
y=data.target
km = KMeans(n_clusters = 3)
km.fit(X,y)
```

```python
predictions = km.predict(data.data)
predictions
```

```python
centroids=km.cluster_centers_
centroids
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,2], c=y, cmap='gist_rainbow')
plt.scatter(centroids[:, 0], centroids[:, 2], c='black', marker='*')
```

```python
data_inertia = km.inertia_
data_inertia
```

```python
number_of_run=[]
inertia=[] 
for i in range (0, 5):
    km = KMeans(n_clusters = i+1)
    km.fit(X,y)
    number_of_run.append(i+1)
    inertia.append(km.inertia_)
inertia
```

```python
import matplotlib.pyplot as plt
plt.plot(number_of_run, inertia, 'ro')
plt.show()
```

```python
inertia = pd.DataFrame(inertia)
plt.figure()
plt.plot(range(0,5),inertia)
plt.xlabel('num of clusters')
plt.ylabel('inertia')
plt.show
```
