```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
vehicle =pd.read_csv("dataset_54_vehicle.csv")
corr = vehicle.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(vehicle.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(vehicle.columns)
ax.set_yticklabels(vehicle.columns)
plt.show()
```

```python
y=vehicle['Class']
x=vehicle
del x['Class']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
tree_cv.fit(X_train, y_train)
```

```python
print("best parameters scores: {}".format(tree_cv.best_params_))
print("best tree score {}".format(tree_cv.best_score_))
```

```python
from sklearn import tree
new_clssfr = tree.DecisionTreeClassifier(max_depth = 5, max_features=4, criterion='entropy').fit(X_train, y_train)
y_pred = new_clssfr.predict(X_test)
y_pred
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
